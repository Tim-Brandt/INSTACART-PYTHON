# Step 2 - Create a bar chart, histogram, scatterplot &  line chart

Bar Chart

```python
#bar chart for loyalty flag customers
bar_loyalty_flag = df_ords_prods_merge['loyalty_flag'].value_counts().plot.bar()
plt.ylabel("# of Orders", fontsize=13)
plt.xlabel("Loyalty Bucket", fontsize=13)  
plt.title("Number of Orders by Customer Loyalty", fontsize=15)
```

```python
#save bar chart
bar_loyalty_flag.figure.savefig(os.path.join(path, 'Analysis','Visualizations', 'loyalty_bar.png'))
```

Histogram

```python
#create histogram
hist_orders_hour_of_day = df_ords_prods_merge['order_hour_of_day'].plot.hist(bins = 24)
plt.xlabel("Hour of Day", fontsize=15)
plt.ylabel("Order Frequency",fontsize=15)
plt.title("Orders Per Hour of Day", fontsize=15)
```

```python
#save histogram
hist_orders_hour_of_day.figure.savefig(os.path.join(path, 'Analysis','Visualizations', 'hist_order_hour_of_day.png'))
```

Scatterplot

```python
#7 scatterplot of age and income
scatter_age_income = sns.scatterplot(x = 'Age', y = 'income',data = df_ords_prods_merge)
```

```python
#save scatterplot
scatter_age_income.figure.savefig(os.path.join(path, 'Analysis','Visualizations', 'scatter_age_income.png'))
```
